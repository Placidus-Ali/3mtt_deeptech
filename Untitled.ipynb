```python
import zipfile
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.preprocessing import image
```

```python
# Unzip the dataset
with zipfile.ZipFile("boat_dataset", 'r') as zip_ref:
    zip_ref.extractall("boat_data")
```

We unzipped a file named "boat_dataset.zip" into a new folder called "boat_data". This process extracted all the contents (images organized in folders like boat/ and not_boat/) so we can use them in our project. After unzipping, we printed out the list of items inside the boat_data folder to confirm that the dataset was successfully extracted and to see the available categories or files.

```python
# Path to boat images
boat_folder = 'boat_data/boat_data/train'  # adjust if needed

# Get list of image files
image_files = [f for f in os.listdir(boat_folder) if f.endswith(('.jpg', '.png', '.jpeg'))][:20]

# Plot
plt.figure(figsize=(15, 10))
for i, file in enumerate(image_files):
    img_path = os.path.join(boat_folder, file)
    img = Image.open(img_path)

    plt.subplot(4, 5, i + 1)  # 4 rows, 5 columns = 20 images
    plt.imshow(img)
    plt.axis('off')
    plt.title(f"Image {i+1}")

plt.tight_layout()
plt.show()
```

We want to show 20 images from our data. So we set the path to a folder containing training images and selected the first 20 image files. Then, we used a loop to open and display each image in a grid of 4 rows and 5 columns using matplotlib. This allowed us to quickly visualize a sample of the dataset to ensure the images were loaded correctly and give us an idea of what the model will be trained on.

```python
# Set up image generator with rescaling
datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)

# Image input size (depends on model later)
IMG_SIZE = (128, 128)
BATCH_SIZE = 32

# Training and validation datasets
train_gen = datagen.flow_from_directory(
    "boat_data",
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary',
    subset='training'
)

val_gen = datagen.flow_from_directory(
    "boat_data",
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary',
    subset='validation'
)
```

We prepared our image data for training using ImageDataGenerator, which helps automatically load, resize, and normalize the images by scaling pixel values between 0 and 1. We also split the dataset into 80% for training and 20% for validation. Then, we told the generator to look inside the boat_data folder, resize each image to 128×128 pixels, group them in batches of 32, and treat it as a binary classification problem (i.e., Boat vs Not Boat). This setup allows us to efficiently feed the images into our model during training and evaluation.

```python
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    MaxPooling2D(2, 2),

    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),

    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')  # Binary classification
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.summary()
```

```python
history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=2
)
```

```python
# Load and preprocess a test image
img_path = "boat_data/image1.jpg"  # change this
img = image.load_img(img_path, target_size=IMG_SIZE)
img_array = image.img_to_array(img) / 255.0
img_array = np.expand_dims(img_array, axis=0)

prediction = model.predict(img_array)

if prediction[0][0] > 0.5:
    print("Prediction: Boat")
else:
    print("Prediction: Not Boat")
```
