KDE Plot - Age vs. Heart Attack Outcome

● Modify the KDE plot to include shade=True, adjust bw_adjust for better smoothing, and use a different palette for better contrast.

● Add a vertical line indicating the median age of patients who had a heart attack.

● Title: "Age Distribution by Heart Attack Outcome"

● X-axis Label: "Age (Years)"

● Y-axis Label: "Density"

```python
# Set figure size and style
plt.figure(figsize=(12, 6))
sns.set_theme(style="darkgrid")

# KDE Plot with enhanced customization
ax = sns.kdeplot(
    data=df, x="Age", hue="Heart_Attack_Outcome",
    fill=True, common_norm=False,
    palette=["blue", "red"], alpha=0.6, linewidth=2.5,
    bw_adjust=0.8,  # Adjust bandwidth for smoother distribution
)

# Overlay a vertical line at the median age for each outcome group
median_no_attack = df[df["Heart_Attack_Outcome"] == 0]["Age"].median()
median_attack = df[df["Heart_Attack_Outcome"] == 1]["Age"].median()

plt.axvline(median_no_attack, color="blue", linestyle="--", linewidth=2, label=f"Median (No Attack): {median_no_attack}")
plt.axvline(median_attack, color="red", linestyle="--", linewidth=2, label=f"Median (Attack): {median_attack}")

# Title and Labels
plt.title("Age Distribution by Heart Attack Outcome", fontsize=15,
          fontweight="bold")
plt.xlabel("Age (Years)", fontsize=13)
plt.ylabel("Density", fontsize=13)

# Custom Legend
handles, labels = ax.get_legend_handles_labels()
plt.legend(handles, labels, loc="upper right", fontsize=12)

# Add grid for better readability
plt.grid(True, linestyle="--", alpha=0.4)

# Show plot
plt.savefig("KDPlot")
plt.show()
```

This KDE(Kernel Density Estimate) plot visualizes the relationship between Age and the presence of Heart Disease. The two curves represent the probability distribution of age for individuals with and without heart disease. It shows how the density of ages differs between these two groups

Heatmap - Correlation of Risk Factors
● Generate a correlation matrix heatmap.

● Adjust annotations, color map, and mask half the heatmap for better readability.

● Ensure the most correlated features are highlighted distinctly.

● Title: "Correlation of Heart Attack Risk Factors"

● X-axis Label & Y-axis Label: Feature Names

```python
# Compute correlation matrix
correlation_matrix = df.corr(numeric_only=True)

# Mask the upper triangle
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# Set up the figure
plt.figure(figsize=(14, 6))

# Generate a heatmap with advanced settings
sns.heatmap(
    correlation_matrix, mask=mask,
    annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.7,
    vmin=-1, vmax=1, cbar_kws={'shrink': 0.8}
)

# Title and display
plt.title("Correlation of Heart Attack Risk Factors", fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.savefig("Corr")
plt.show()
```

This is a Correlation Heatmap of "Heart Attack Risk Factors." It uses a color gradient (blue for negative, red for positive) to show the strength and direction of linear relationships between pairs of variables. Numbers within cells represent correlation coefficients. A "mask" hides redundant upper-triangle values. It highlights which factors are positively or negatively related, and to what extent, to the "Heart_Attack_Outcome" (e.g., Age and obesity indes have a strong correlation)

Boxplot - Cholesterol Levels by Heart Attack Outcome

● Modify the boxplot to include hue="Gender" to compare cholesterol levels across males and females.

● Adjust outlier visibility and tweak whiskers to better visualize cholesterol distribution.

● Title: "Cholesterol Levels Distribution by Heart Attack Outcome and Gender"

● X-axis Label: "Heart Attack Outcome (0 = No, 1 = Yes)"

● Y-axis Label: "Cholesterol Level (mg/dL)

```python
plt.figure(figsize=(12, 6))

# Correct boxplot
sns.boxplot(
    data=df,
    x="Heart_Attack_Outcome",
    y="Cholesterol_Level",
    hue="Gender",
    palette="coolwarm",
    showfliers=False,
    whis=1.5
)

# Labels and title
plt.title("Cholesterol Levels Distribution by Heart Attack Outcome and Gender", fontsize=14, fontweight="bold")
plt.xlabel("Heart Attack Outcome (0 = No, 1 = Yes)")
plt.ylabel("Cholesterol Level (mg/dL)")
plt.legend(title="Gender")
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.savefig("Boxplot")
plt.show()
```

The plot shows how having a heart attack is associated with significant higher cholesterol level, and the difference in cholesterol levels between heart attack outcomes appears consistent across both genders. Gender itself does not seem to be a major factor in cholesterol distribution within each outcome group

Pairplot - Multiple Risk Factors & Outcome

● Create a pairplot including Cholesterol_Level, Obesity_Index, LDL_Level, and Heart_Attack_Outcome.

● Adjust markers, increase plot size, and set diag_kind="kde" for better insight.

● Title: "Pairwise Comparison of Key Risk Factors"

● X-axis Label & Y-axis Label: Feature Names

```python
# Set up the figure
plt.figure(figsize=(14, 6))
# Create a pairplot with additional styling
pairplot = sns.pairplot(
    df,
    vars=["Cholesterol_Level", "Obesity_Index", "LDL_Level"],
    hue="Heart_Attack_Outcome",
    palette={0: "blue", 1: "red"},  # Explicit mapping for better readability
    markers=["o", "s"],  # Different markers for classes
    diag_kind="kde",  # Use KDE plots for diagonal
    plot_kws={"alpha": 0.7}  # Add transparency for better visualization
)

# Add a title
plt.subplots_adjust(top=0.95)
pairplot.fig.suptitle("Pairwise Comparison of Key Risk Factors", fontsize=14, fontweight='bold')
plt.savefig("Pairplot")
plt.show()
```

This pair plot visualizes relationships among Cholesterol_Level, Obesity_Index, and LDL_Level, colored by Heart_Attack_Outcome (blue=No, red=Yes).

Diagonal plots show distributions: individuals with heart attacks (red) tend to have higher Cholesterol_Level and LDL_Level, and slightly higher Obesity_Index. Off-diagonal scatter plots reveal pairwise relationships, highlighting that higher values in these risk factors are generally more associated with the "Heart Attack" outcome (red points)

Violin Plot - Blood Pressure by Heart Attack Outcome

● Create a violin plot to compare Blood_Pressure_Systolic and Blood_Pressure_Diastolic across heart attack outcomes.

● Adjust split=True to better visualize distributions.

● Title: "Blood Pressure Distribution by Heart Attack Outcome"

● X-axis Label: "Heart Attack Outcome (0 = No, 1 = Yes)"

● Y-axis Label: "Blood Pressure (mmHg)"

```python
# Reshape the data to combine systolic and diastolic into one column
bp_df = df.melt(
    id_vars="Heart_Attack_Outcome",
    value_vars=["Blood_Pressure_Systolic", "Blood_Pressure_Diastolic"],
    var_name="Blood_Pressure_Type",
    value_name="Blood_Pressure"
)

# Set up the figure
plt.figure(figsize=(10, 6))

# Create the violin plot with split=True and proper hue
ax = sns.violinplot(
    x="Heart_Attack_Outcome",
    y="Blood_Pressure",
    hue="Blood_Pressure_Type",
    data=bp_df,
    palette="coolwarm",
    split=True,
    inner="quartile",
    bw_adjust=0.2
)

plt.title("Blood Pressure Distribution by Heart Attack Outcome", fontsize=14, fontweight="bold")
plt.xlabel("Heart Attack Outcome (0 = No, 1 = Yes)", fontsize=12)
plt.ylabel("Blood Pressure (mmHg)", fontsize=12)
plt.xticks([0, 1], ["No Heart Attack", "Heart Attack"])
plt.legend(title="Blood Pressure Type")
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.savefig("Violinplot")
plt.show()
```

This violin plot illustrates the distribution of Systolic and Diastolic Blood Pressure for individuals with and without a Heart Attack Outcome.

For both Systolic (blue) and Diastolic (orange) pressure, the distributions are shifted to higher values for those who had a heart attack ("Heart Attack" group) compared to those who did not. This indicates that individuals who experience heart attacks generally have higher blood pressure readings, both systolic and diastolic.

Countplot - Smoking and Heart Attack Outcome

● Generate a countplot to visualize the number of smokers and non-smokers across heart attack outcomes.

● Use a hue="Gender" parameter to compare distributions by gender.

● Title: "Smoking Status and Heart Attack Outcome by Gender"

● X-axis Label: "Smoking Status (Yes/No)"

● Y-axis Label: "Count"

```python
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

# Create the countplot
ax = sns.countplot(
    data=df,
    x="Smoking_Status",
    hue="Gender",
    palette="Set2"
)

plt.title("Smoking Status and Heart Attack Outcome by Gender", fontsize=14, fontweight="bold")
plt.xlabel("Smoking Status (Yes/No)", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.legend(title="Gender")
plt.savefig("Countplot")
plt.show()
```

This countplot shows the distribution of smoking status (Yes/No) by Gender (Female/Male) in the dataset.

It indicates that the majority of individuals in the dataset are non-smokers, with significantly higher counts for both females (teal) and males (orange) in the "No" smoking status category. Among non-smokers, the counts for females and males are almost identical. Similarly, among smokers ("Yes"), the counts for both genders are also very close, though much lower than non-smokers
