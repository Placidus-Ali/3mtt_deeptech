```python
def has_special_characters(series):
    special_chars = [">", "<", "!", "'"]
    for char in special_chars:
        try:
            if series.astype(str).str.contains(char, regex=False).any():
                return True
        except:
            pass 
    return False

columns_with_special_chars = []
for col in job_df.columns:
    if has_special_characters(job_df[col]):
        columns_with_special_chars.append(col)

print("Columns are:", columns_with_special_chars)
```

Bar Chart of Average GPA by Grade Class:

● Check for any missing or inconsistent data and handle them accordingly. Group the data by GradeClass and calculate the average GPA for each grade class.

● Create a bar chart displaying the average GPA for each grade class.

● Use a colormap to color the bars based on the GPA values. Choose a color map that reflects the gradient from low to high GPA.

● Add annotations to display the exact GPA values on top of the bars.

● Make sure the chart has an appropriate title, and labels for the x and y axes.

```python
# Prepare data
avg_gpa_by_grade = student_df.groupby('GradeClass')['GPA'].mean().reset_index()
x = avg_gpa_by_grade['GradeClass'].values
y = avg_gpa_by_grade['GPA'].values

# Normalize GPA values for colormap scaling (between 0 and 1)
norm = plt.Normalize(y.min(), y.max())
cmap = plt.cm.viridis
colors = cmap(norm(y))

# Create figure and axis
fig, ax = plt.subplots(figsize=(10,8))

# Draw bars manually using a loop
bar_width = 0.6
for i, (grade, gpa) in enumerate(zip(x, y)):
    ax.add_patch(plt.Rectangle((i - bar_width/2, 0), bar_width, gpa, color=colors[i]))
    # Add annotation on top of the bar
    ax.text(i, gpa + 0.03, f'{gpa:.2f}', ha='center', va='bottom', fontsize=10)
    # Here we're looping over each grade class, drawing a bar as a rectangle with height = GPA, 
    #and then placing a text label right above it to show the GPA value

# Set labels and title
ax.set_xlabel('Grade Class')
ax.set_ylabel('Average GPA')
ax.set_title('Average GPA by Grade Class')

# Set limits
ax.set_xlim(-0.5, len(x) - 0.5)
ax.set_ylim(0, y.max() + 0.5)

#plt.savefig('Average')
plt.show()
```

Scatter Matrix for Study Time, Absences, and GPA:

● Create a scatter matrix for the variables StudyTimeWeekly, Absences, and GPA.

● Set the diagonal to histograms to show the distribution of each variable.

● Use a color that stands out for the points, and add titles and grid lines to improve readability.

```python
selected_df = student_df[['StudyTimeWeekly', 'Absences', 'GPA']]

# Set a color
scatter_color = "#1f77b4"  

# Create the pair plot
sns.set(style="whitegrid")  # Add grid lines for better readability
pair_plot = sns.pairplot(
    selected_df,
    diag_kind="hist", 
    plot_kws={'color': scatter_color, 's': 40, 'alpha': 0.6}  # style of scatter points
)

# Set title to the figure
pair_plot.fig.suptitle("Scatter Matrix of Study Time, Absences, and GPA", fontsize=14)
pair_plot.fig.tight_layout()
pair_plot.fig.subplots_adjust(top=0.92)  # Adjust to make room for title
plt.savefig('Scatter')
plt.show()
```

[1,1] This indicates a highly right-skewed distribution. The vast majority of students report very low weekly study times.

[1,2] This shows no discernible linear correlation. The low study time values are present regardless of absence count. The spread is more horizontal for the main cluster, indicating that study time tends to be low, independent of how many absences a student has.

[1,3] This also shows no clear linear correlation. It suggests that for the majority of students (those with low study time), GPA can vary widely. There's no obvious trend where higher study time consistently leads to higher GPA.

[2,1] Reflects the same lack of linear correlation. The distribution shows that Absences are varied, primarily for students who study very little.

[2,2] This suggests that Absences are spread out across their range, with some specific counts being more common than others, but without the extreme skew observed in StudyTimeWeekly

[2,3] This indicates a clear negative linear correlation. Higher GPAs are generally associated with fewer absences, and lower GPAs are associated with more absences.

[3,1] There is no strong linear correlation. Students who study very little can still have a wide range of GPAs.

[3,2] This clearly shows a strong negative linear correlation. More absences are associated with lower GPAs. 

[3,3] Most students have a GPA in the middle range, with fewer having very low or very high GPAs.

In summary, the scatter matrix efficiently shows individual variable distributions on the diagonal and pairwise relationships (correlations) in the off-diagonal plots. The most significant finding here is the clear negative correlation between Absences and GPA

Box Plot for Study Time Distribution by Grade Class:

● Create a box plot to visualize the distribution of StudyTimeWeekly for each GradeClass.

● Ensure each grade class has a corresponding box, and label them properly.

● Add a grid, axis labels, and a title to the plot.

● Customize the box plot with different colors for each grade class.

```python
# Set plot style
sns.set(style="whitegrid")

# Create a color palette with distinct colors for each GradeClass
unique_grades = sorted(student_df['GradeClass'].unique())
palette = sns.color_palette("Set2", len(unique_grades))  # 'set2' is just an in-built color palette in seaborn

# Create the boxplot
plt.figure(figsize=(10, 6))
ax = sns.boxplot(
    data=student_df,
    x='GradeClass',
    y='StudyTimeWeekly',
    hue= 'GradeClass',
    palette=palette
)

# Add labels and title
plt.xlabel("Grade Class", fontsize=12)
plt.ylabel("Study Time per Week (hours)", fontsize=12)
plt.title("Study Time Distribution by Grade Class", fontsize=14)

# Display the plot
plt.savefig('Boxplot')
plt.show()
```

This shows that GradeClass does not have a strong differentiator for how much students study on a weekly basis, at least within the typical range. The majority of students across all classes fall into a similar pattern of low study hours
