```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("heart_attack_south_africa.csv")
df
```

**Advanced Cholesterol Distribution Plot:**

● Create a Cholesterol Distribution Plot (histogram + KDE curve) to analyze the distribution of cholesterol levels in the dataset.

● Customize the plot with gridlines, labels, and a title.

● Examine the plot and discuss insights related to cholesterol levels.

```python
# pandas for calculations
mean_chol_no_attack = df[df['Heart_Attack_Outcome'] == 0]['Cholesterol_Level'].mean()
mean_chol_attack = df[df['Heart_Attack_Outcome'] == 1]['Cholesterol_Level'].mean()

median_chol_no_attack = df[df['Heart_Attack_Outcome'] == 0]['Cholesterol_Level'].median()
median_chol_attack = df[df['Heart_Attack_Outcome'] == 1]['Cholesterol_Level'].median()

# Set Seaborn style (Seaborn)
sns.set_style("whitegrid")  # This applies a polished white-grid background to the plot

# Create figure and axis (Matplotlib)
plt.figure(figsize=(10, 6))  # Set figure size using Matplotlib

# Seaborn histogram with KDE (Seaborn)
sns.histplot(df[df['Heart_Attack_Outcome'] == 0]['Cholesterol_Level'], kde=True, color='green', bins=30, alpha=0.6, label='No Heart Attack')
sns.histplot(df[df['Heart_Attack_Outcome'] == 1]['Cholesterol_Level'], kde=True, color='red', bins=30, alpha=0.6, label='Heart Attack')

# Add mean and median lines for each group (Matplotlib)
plt.axvline(mean_chol_no_attack, color='green', linestyle='--', linewidth=2, label=f"Mean Cholesterol (No Attack): {mean_chol_no_attack:.2f}")  # Mean line for No Heart Attack
plt.axvline(mean_chol_attack, color='red', linestyle='--', linewidth=2, label=f"Mean Cholesterol (Heart Attack): {mean_chol_attack:.2f}")  # Mean line for Heart Attack

plt.axvline(median_chol_no_attack, color='green', linestyle='-', linewidth=2, label=f"Median Cholesterol (No Attack): {median_chol_no_attack:.2f}")  # Median line for No Heart Attack
plt.axvline(median_chol_attack, color='red', linestyle='-', linewidth=2, label=f"Median Cholesterol (Heart Attack): {median_chol_attack:.2f}")  # Median line for Heart Attack

# Matplotlib Customization (Matplotlib)
plt.title("Cholesterol Level Distribution Based on Heart Attack Outcome", fontsize=15, fontweight="bold")  # Set title
plt.xlabel("Cholesterol Level", fontsize=13)  # X-axis label
plt.ylabel("Frequency", fontsize=13)  # Y-axis label
plt.legend()  # Display the legend for mean and median lines
plt.grid(True, linestyle="--", alpha=0.7)  # Add grid with dashed lines and transparency

# Show the plot (Matplotlib)
plt.savefig('Hist_KDE')
plt.show()
```

**Observations**

No Heart Attack (Green): Lower cholesterol levels, with a mean around 209.91 and median around 206.00.

Heart Attack (Red): Higher cholesterol levels, with a mean around 234.89 and median around 244.00.

Risk: Higher cholesterol levels are associated with an increased frequency of heart attacks.

**Overlaying Line and Scatter Plots:**

● Overlay a scatter plot and a line plot to show the relationship between two
variables (e.g., age vs. cholesterol levels).

● Customize the appearance of both plots (e.g., color, size, transparency,
etc.).

● State your observations for the relationship between the variables based on
the plots.

```python
# Create subplots with two axes (Matplotlib)
fig, ax = plt.subplots(1, 2, figsize=(14, 6))  # 1 row, 2 columns of plots

# ======== Seaborn Boxplot (First Plot) ========
# Boxplot for Cholesterol Level across Smoking Status (Seaborn)
sns.boxplot(data=df, x="Smoking_Status", y="Cholesterol_Level", ax=ax[0], palette="Set2")

# Matplotlib customization (First Axis)
ax[0].set_title("Cholesterol Levels by Smoking Status", fontsize=13)
ax[0].set_xlabel("Smoking Status", fontsize=12)
ax[0].set_ylabel("Cholesterol Level", fontsize=12)

# ===== Seaborn KDE Plot with Matplotlib Line (Second Plot) =========

# Kernel Density Estimation (KDE) Plot for LDL Level based on Heart Attack Outcome (Seaborn)
sns.kdeplot(data=df, x="LDL_Level", hue="Heart_Attack_Outcome", fill=True, alpha=0.4, ax=ax[1])

# Overlay Matplotlib sine wave for illustration (Matplotlib)
x = np.linspace(df["LDL_Level"].min(), df["LDL_Level"].max(), 100)
y = np.sin(x / max(x) * np.pi) * 20  # Normalizing x and scaling sine wave
ax[1].plot(x, y, color="black", linestyle="--", label="Custom Sine Wave")

# Matplotlib customization (Second Axis)
ax[1].set_title("LDL Levels KDE Plot with Overlay", fontsize=13)
ax[1].set_xlabel("LDL Level", fontsize=12)
ax[1].set_ylabel("Density", fontsize=12)
ax[1].legend()

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.savefig('Scatterplot')
plt.show()
```

The left box plot shows cholesterol levels by smoking status. Both non-smokers and smokers have similar median cholesterol levels (around 220-225), and their interquartile ranges largely overlap. This suggests that smoking status alone doesn't appear to be a strong differentiator of cholesterol levels based on this data. 

The right plot is an LDL levels KDE plot with a sine wave overlay, but without further context on the "Custom Sine Wave"

**Creating Multi-Axes Plots:**
    
● Create a multi-plot figure with Matplotlib’s subplots function.

● In one axis, plot a bar chart showing cholesterol levels by gender.
    
● In another axis, plot a scatter plot showing the relationship between blood pressure and age.

● Add titles, axis labels, and legends to each plot

```python
# Create a figure with two subplots (Matplotlib)
fig, ax = plt.subplots(1, 2, figsize=(14, 6))  # 1 row, 2 columns

# ========== Seaborn Boxplot (First Axis) ======
# Boxplot for Cholesterol Level across Heart Attack Outcome (Seaborn)
sns.boxplot(data=df, x="Heart_Attack_Outcome", y="Cholesterol_Level", ax=ax[0], palette="Set2")

# Matplotlib customization (First Axis)
ax[0].set_title("Cholesterol Levels by Heart Attack Outcome", fontsize=13)
ax[0].set_xlabel("Heart Attack Outcome", fontsize=12)
ax[0].set_ylabel("Cholesterol Level", fontsize=12)

# ====== Matplotlib Scatter Plot (Second Axis) ========
# Scatter plot for LDL Level vs Triglycerides Level (Matplotlib)
sc = ax[1].scatter(df["LDL_Level"], df["Triglycerides_Level"], c=df["Heart_Attack_Outcome"], cmap="coolwarm", alpha=0.7)

# Matplotlib customization (Second Axis)
ax[1].set_title("LDL vs Triglycerides Level", fontsize=13)
ax[1].set_xlabel("LDL Level", fontsize=12)
ax[1].set_ylabel("Triglycerides Level", fontsize=12)

# Add colorbar for better visualization
cbar = fig.colorbar(sc, ax=ax[1])
cbar.set_label("Heart Attack Outcome")

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.savefig('Multi_Axes')
plt.show()
```

The box plot shows cholesterol levels by heart attack outcome (0 = No Heart Attack, 1 = Heart Attack). It indicates that individuals who had a heart attack (group 1) generally have higher cholesterol levels, with a higher median and interquartile range compared to those who did not (group 0).

The scatter plot visualizes LDL vs. Triglycerides level. It appears to show a scattered distribution of points with no clear linear correlation between LDL and Triglycerides levels. 

**Combining Seaborn’s FacetGrid and Matplotlib:**

● Use Seaborn’s FacetGrid to create a grid of subplots, segmented by a categorical feature (e.g., smoking status or diabetes status).
                                                                                            
● Inside the grid, plot a scatter plot or box plot of the heart disease outcomes based on different categorical features.
    
● Customize the appearance of the FacetGrid using Matplotlib (e.g., change axis limits, add titles).

```python

# Create a FacetGrid to plot scatter plots for different Smoking Status categories
g = sns.FacetGrid(df, col="Smoking_Status", height=4, aspect=0.8)

# Map Seaborn scatter plots onto the grid (Seaborn)
g.map(sns.scatterplot, "LDL_Level", "HDL_Level", alpha=0.6)

# ====Matplotlib Customization (Annotations) =========
# Add custom horizontal reference lines (Matplotlib)
for ax in g.axes.flat:
    ax.axhline(y=50, color="red", linestyle="--", label="Threshold HDL Level")  # Reference line

# Add a title to the entire figure (Matplotlib)
g.fig.suptitle("LDL vs HDL Levels Across Smoking Status", y=1.02, fontsize=14)

# Show the final plot
plt.savefig('Facegrid')
plt.show()
```

This plot suggests that smoking status does not visually differentiate the relationship or distribution of LDL and HDL levels in this dataset

**Advanced Bar Chart with Seaborn and Matplotlib:**
    
● Create an advanced bar chart comparing cholesterol levels across different categories (e.g., gender, smoking status). 

● Add custom error bars to the bars and include data labels on the bars.

● Use Matplotlib for further customization, such as adjusting bar width and colors.

```python
# Check for any missing values in the relevant columns
df.dropna(subset=['Smoking_Status', 'Heart_Attack_Outcome'], inplace=True)

# Create a figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# Create the Seaborn barplot
sns.countplot(x="Smoking_Status", hue="Heart_Attack_Outcome", data=df, palette="Set1", ax=ax)

# Customize the plot using Matplotlib
ax.set_title("Heart Attack Outcome by Smoking Status", fontsize=16)
ax.set_xlabel("Smoking Status", fontsize=12)
ax.set_ylabel("Count of Heart Attack Outcomes", fontsize=12)

# Add grid lines to make the plot more readable
ax.grid(True, linestyle='--', alpha=0.7)

# Annotate each bar with the count of occurrences
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height}', xy=(p.get_x() + p.get_width() / 2., height),
                xytext=(0, 5), textcoords="offset points",
                ha='center', va='bottom', fontsize=12, color='black')

# Add a legend with a custom location
ax.legend(title="Heart Attack Outcome", loc='upper right', fontsize=10)

# Display the plot
plt.tight_layout()
plt.savefig('Advanced Bar Chart')
plt.show()
```

This bar chart displays heart attack outcomes (0 = No Heart Attack, 1 = Heart Attack) categorized by smoking status. 

A significantly higher number of non-smokers (35,630) did not experience a heart attack compared to the number who did (34,359).

Among smokers, a larger proportion experienced a heart attack (24,373) than did not (5,638).

This suggests a strong association where smoking dramatically increases the likelihood of a heart attack outcome.
