```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.arange(1, 6)
y = np.array([1, 4, 2, 5, 3])
```

```python
# 1. Line Plot
plt.plot(x, y)
plt.title("Line Plot")
plt.show()
```

```python
# 2. Bar Chart
plt.bar(x, y)
plt.title("Bar Chart")
plt.show()
```

```python
# 3. Horizontal Bar Chart
plt.barh(x, y)
plt.title("Horizontal Bar Chart")
plt.show()
```

```python
# 4. Box Plot
plt.boxplot(y)
plt.title("Box Plot")
plt.show()
```

```python
# 5. Pie Chart
plt.pie(y, labels=x)
plt.title("Pie Chart")
plt.show()
```

```python
# 6. Area Plot (fill_between)
plt.fill_between(x, y)
plt.title("Area Plot")
plt.show()

```

```python
# Sample DataFrame
df = sns.load_dataset("tips")
df
```

```python
# 1. Histogram
sns.histplot(df['total_bill'])
plt.title("Histogram")
plt.show()
```

```python
# 2. KDE Plot
sns.kdeplot(df['total_bill'])
plt.title("KDE Plot")
plt.show()
```

```python
# 3. Scatter Plot
sns.scatterplot(x="total_bill", y="tip", data=df)
plt.title("Scatter Plot")
plt.show()
```

```python
# 4. Line Plot
sns.lineplot(x="size", y="tip", data=df)
plt.title("Line Plot")
plt.show()
```

```python
# 5. Bar Plot
sns.barplot(x="day", y="total_bill", data=df)
plt.title("Bar Plot")
plt.show()
```

```python
# 6. Count Plot
sns.countplot(x="day", data=df)
plt.title("Count Plot")
plt.show()
```

```python
# 7. Box Plot
sns.boxplot(x="day", y="total_bill", data=df)
plt.title("Box Plot")
plt.show()
```

```python
# 8. Heatmap
# mask = np.triu(np.ones_like(corr, dtype=bool))

df1 = df[['total_bill', 'tip', 'size']]
corr = df1.corr()
sns.heatmap(corr, annot=True)
plt.title("Heatmap")
plt.show()
```

```python
# 9. Pairplot
sns.pairplot(df[['total_bill', 'tip', 'size']])
plt.title("Pairplot")
plt.show()
```
